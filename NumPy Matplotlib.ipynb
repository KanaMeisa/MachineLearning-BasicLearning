# Matplotlib
是Python的绘图库，它可以和NumPy、PyQt 或 wxPython 一起使用

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family']=['Source Han Sans CN']                    # 默认情况下不支持中文，需要手动添加字体

x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)
plt.title("测试")
plt.xlabel("X Axis Caption")
plt.ylabel("Y Axis Caption")
plt.plot(x, y, 'r')
plt.show()
```

## bar()
pyplot子模块提供bar()来生成条形图

```python
x = [5, 8, 10]
y = [12, 16, 6]
x2 = [6, 9, 11]
y2  = [6, 15, 7]
plt.bar(x, y, color='r',align='center')
plt.bar(x2, y2, align='center')
plt.title("Bar Graph")
plt.ylabel("Y")
plt.xlabel("X")
plt.show()
```
